```python
days=tips_df.day.unique()
print(days)
```

```python
tips_df['day'].count()
```

```python
tips_df[tips_df['day']=='Sun']['day'].count()
```

```python


day_data={};
for eachday in days:
  day_data[eachday]=[tips_df[tips_df['day']==eachday]['day'].count()]

print(day_data)
```

```python
day_df = pd.DataFrame(day_data)
print(day_df)
```

```python
sns.barplot(day_df)
```

```python
tips_series=tips_df.groupby('day')['day'].count()
sns.barplot(x=tips_series.index,y=tips_series.values)
```

RelationalPlots
